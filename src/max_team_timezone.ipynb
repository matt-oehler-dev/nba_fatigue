```python
import pandas as pd
from nba_api.stats.static import teams

# Get all NBA teams
nba_teams = pd.DataFrame(teams.get_teams())

# Add the time zone for each team manually
team_time_zones = {
    'ATL': 'America/New_York', 'BOS': 'America/New_York', 'CLE': 'America/New_York', 
    'NOP': 'America/Chicago', 'CHI': 'America/Chicago', 'DAL': 'America/Chicago',
    'DEN': 'America/Denver', 'GSW': 'America/Los_Angeles', 'HOU': 'America/Chicago',
    'LAC': 'America/Los_Angeles', 'LAL': 'America/Los_Angeles', 'MIA': 'America/New_York',
    'MIL': 'America/Chicago', 'MIN': 'America/Chicago', 'BKN': 'America/New_York',
    'NYK': 'America/New_York', 'ORL': 'America/New_York', 'IND': 'America/New_York',
    'PHI': 'America/New_York', 'PHX': 'America/Phoenix', 'POR': 'America/Los_Angeles',
    'SAC': 'America/Los_Angeles', 'SAS': 'America/Chicago', 'OKC': 'America/Chicago',
    'TOR': 'America/Toronto', 'UTA': 'America/Denver', 'MEM': 'America/Chicago',
    'WAS': 'America/New_York', 'DET': 'America/New_York', 'CHA': 'America/New_York'
}

utc_offsets = {
    'America/New_York': -4,
    'America/Chicago': -5,
    'America/Denver': -6,
    'America/Los_Angeles': -8,
    'America/Phoenix': -7,
    'America/Toronto': -4,
    'America/Utah': -7
}

# Add the time zones to the DataFrame
nba_teams['timezone'] = nba_teams['abbreviation'].map(team_time_zones)
nba_teams['time_diff'] = nba_teams['timezone'].map(utc_offsets) + 4

# Save the DataFrame to a CSV file
nba_teams.to_csv('../data/team_info_with_timezone.csv', index=False)
nba_teams
```
