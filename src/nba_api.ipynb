### Player Box Scores


```python
from nba_api.stats.endpoints import leaguegamefinder

gamefinder = leaguegamefinder.LeagueGameFinder(
    player_or_team_abbreviation="P",
    season_type_nullable="Regular Season",
    season_nullable="2022-23",
)

df = gamefinder.get_data_frames()[0]
print(df.shape)
df.head()
```

##


```python
# make team game key
df['TEAM_GAME_KEY'] = df['TEAM_ABBREVIATION'] + '_' + df['GAME_ID']
df['GAME_DATE'] = pd.to_datetime(df['GAME_DATE'])

# mark as guard or not guard
df['IS_GUARD'] = np.where(df['PLAYER_ID'].isin(guards['PLAYER_ID']), 1, 0)


print(df.shape)
df.sample(2)
```

```python
# # temporary constraint
# df=df[
#     df['TEAM_ABBREVIATION'].isin(["POR"])
# ]
# print(df.shape)
# df.head()
```

```python
# make a lookup table that has a row for each game for each team
# so total rows = total games played * 2 
# since there is one row for each team per game
team_games = (
        df.pivot_table(
            index=["TEAM_ID","TEAM_ABBREVIATION", "TEAM_GAME_KEY", 'GAME_DATE', 'MATCHUP', "WL"],
            values=["PLAYER_ID"],
            aggfunc=pd.Series.nunique,
        )
        .reset_index()
        .drop(['PLAYER_ID'], axis=1)
        .sort_values(by=["TEAM_ID",'GAME_DATE'])
)
team_games["TEAM_GAME_ID"] = team_games.groupby(["TEAM_ABBREVIATION"]).cumcount() + 1

print(team_games.shape)
# team_games.tail(25)
team_games.sample(10)
```

```python
temp = team_games[
    team_games['TEAM_ABBREVIATION'] == "POR"
]
print(temp.shape)
temp.head(10)
```

```python
team_games['TEAM_GAME_ID'].max()
```

```python
# make a table of players (1 row per player)
# note that this only includes players that have played in games
team_players = (
    df.pivot_table(
        index=["TEAM_ID","TEAM_ABBREVIATION", 'PLAYER_NAME', "PLAYER_ID"],
        values=['IS_GUARD'],
        aggfunc=max
    )
    .reset_index()
    # .drop("TEAM_GAME_KEY", axis=1)
)

print(team_players.shape)
team_players.head(30)
```

```python
# now join to get a table that has a row for each playe
tp_cols = [
    "TEAM_ID",
    # "TEAM_ABBREVIATION",
    "PLAYER_NAME",
    "PLAYER_ID",
    'IS_GUARD'
]
game_players = pd.merge(
    team_games,
    team_players[tp_cols],
    # on="TEAM_ID",
    left_on="TEAM_ID",
    right_on="TEAM_ID",
    how="left",
)

print(game_players.shape)

game_players.tail(25)
```

```python
df.columns
# df.head()
```

```python
cols = [
    # "SEASON_ID",
    "PLAYER_ID",
    "TEAM_GAME_KEY",
    # "PLAYER_NAME",
    # "TEAM_ID",
    # "TEAM_ABBREVIATION",
    # "TEAM_NAME",
    # "GAME_ID",
    # "GAME_DATE",
    # "MATCHUP",
    # "WL",
    "MIN",
    "PTS",
    # "FGM",
    # "FGA",
    # "FG_PCT",
    # "FG3M",
    # "FG3A",
    # "FG3_PCT",
    # "FTM",
    # "FTA",
    # "FT_PCT",
    # "OREB",
    # "DREB",
    # "REB",
    # "AST",
    # "STL",
    # "BLK",
    # "TOV",
    # "PF",
    # "PLUS_MINUS",
    # "IS_GUARD",
]

final = pd.merge(
    game_players,
    df[cols],
    left_on=["PLAYER_ID", "TEAM_GAME_KEY"],
    right_on=["PLAYER_ID", "TEAM_GAME_KEY"],
    how="left",
)

# make new columns
final['IS_GAME_PLAYED'] = np.where(~final['MIN'].isna(), 1, 0)

final.fillna(0, inplace=True)

final['cum_points'] = final.groupby(["PLAYER_ID"])['PTS'].cumsum()
final['TOTAL_GAMES_PLAYED'] = final.groupby(["PLAYER_ID"])['IS_GAME_PLAYED'].cumsum().fillna(0)
final['cum_ppg'] = (final.groupby(["PLAYER_ID"])['PTS'].cumsum() / final['TOTAL_GAMES_PLAYED']).fillna(0)


# save to file
final.to_csv("player_box_scores.csv", index=False)

print(final.shape)
# final.head(17)
```

```python
master = pd.merge(
    temp[['team_game_key', 'team_game_id']],
    df,
    left_on='team_game_key',
    right_on='team_game_key',
    how='outer'
).sort_values(by=['TEAM_ID', 'GAME_DATE', "PLAYER_ID"])
print(df.shape)
print(master.shape)

master.head(2)
```

```python
# make cumulative ppg
master['cum_points'] = master.groupby(["PLAYER_ID"])['PTS'].cumsum()
master['cum_ppg'] = master.groupby(["PLAYER_ID"])['PTS'].cumsum() / master['team_game_id']
# df['cumsum_col'] = df.groupby(['col1'])['col2'].cumsum()


master[master['PLAYER_NAME'] == "Ja Morant"]
```

```python
# get the top 5 scorers
top_scorers = (
    master.pivot_table(
        index=["PLAYER_ID", "PLAYER_NAME", "TEAM_NAME"],
        values=["PTS", "AST"],
        aggfunc="sum",
    )
    .sort_values(by="PTS", ascending=False)
    .reset_index()
    .head(5)
)

top_scorers
```

```python
# get the game data for each of the top scorers
temp = df[
    (df["PLAYER_ID"].isin(top_scorers["PLAYER_ID"]))
    & (df["PLAYER_ID"].isin(guards["PLAYER_ID"]))
]
# temp['GAME_DATE'] = pd.to_datetime(temp['GAME_DATE'])

temp = temp.sort_values(by=["PLAYER_ID", "GAME_ID"])
temp["GAME_NUM"] = temp.groupby("PLAYER_ID").cumcount() + 1
temp["PPM"] = temp["PTS"] / temp["MIN"]

print(temp.shape)
temp
```

```python
# plot
ymin = temp["PTS"].min() - 5
ymax = temp["PTS"].max() + 5
xmin = temp["GAME_NUM"].min() - 0.05
xmax = temp["GAME_NUM"].max() + 3

sns.lineplot(x="GAME_NUM", y="PTS", hue="PLAYER_NAME", data=temp, marker="o")
plt.title("Highest Scoring NBA Guards 2022-23")
plt.ylim(ymin, ymax)
plt.xlim(xmin, xmax)
plt.grid()
plt.show()
```
